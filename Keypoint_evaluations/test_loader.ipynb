```python
import rosbag
bag = rosbag.Bag(bag_name)
indx=0
for topic, msg, t in bag.read_messages(topics=['/ouster_points']):
    points = []
    for point in pc2.read_points(msg, skip_nans=True):
        points.append([point[0],point[1],point[2] ])   
    points = np.array(points)
    #print(points.shape)
    frame_name = synth_data_dir_seq+"/"+ '{0:06d}.npy'.format(indx)
    np.save(frame_name, points)
    indx+=1
    print(frame_name)
    #break
bag.close()
```
